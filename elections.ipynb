```python
pres2002 = pd.read_csv("./donnees/élections/pres2002comm.csv")
pres2007 = pd.read_csv("./donnees/élections/pres2007comm.csv")
pres2017 = pd.read_csv("./donnees/élections/pres2017comm.csv")
pres2022 = pd.read_csv("./donnees/élections/pres2022comm.csv")
```

```python
pres2002['codecommune'] = pres2002['codecommune'].astype('string').str.zfill(5)
pres2007['codecommune'] = pres2007['codecommune'].astype('string').str.zfill(5)
pres2017['codecommune'] = pres2017['codecommune'].astype('string').str.zfill(5)
pres2022['codecommune'] = pres2022['codecommune'].astype('string').str.zfill(5)
```

```python
vote_2002 = pd.merge(communes_2024, pres2002, left_on='code', right_on='codecommune',how='inner').drop(columns=['dep','nomdep','codecommune','nomcommune'])
vote_2007 = pd.merge(communes_2024, pres2007, left_on='code', right_on='codecommune',how='inner').drop(columns=['dep','nomdep','codecommune','nomcommune'])
vote_2017 = pd.merge(communes_2024, pres2017, left_on='code', right_on='codecommune',how='inner').drop(columns=['dep','nomdep','codecommune','nomcommune'])
vote_2022 = pd.merge(communes_2024, pres2022, left_on='code', right_on='codecommune',how='inner').drop(columns=['dep','nomdep','codecommune','nomcommune'])
```

```python
diff_melenchon_2017_2022 = vote_2022[['code','nom','geometry','pvoixMELENCHON']]
diff_melenchon_2017_2022 = diff_melenchon_2017_2022.merge(vote_2017[['code', 'pvoixMELENCHON']].rename(columns={'pvoixMELENCHON': 'pvoixMELENCHON_2017'}),on='code')
```

```python
diff_melenchon_2017_2022['diff'] = diff_melenchon_2017_2022['pvoixMELENCHON'] - diff_melenchon_2017_2022['pvoixMELENCHON_2017']
```

```python
diff_melenchon_2017_2022[diff_melenchon_2017_2022['nom']=='Mellionnec']
```

```python
diff_mlepen_2017_2022 = vote_2022[['code','nom','geometry','pvoixMLEPEN']]
diff_mlepen_2017_2022 = diff_mlepen_2017_2022.merge(vote_2017[['code', 'pvoixMLEPEN']].rename(columns={'pvoixMLEPEN': 'pvoixMLEPEN_2017'}),on='code')
```

```python
diff_mlepen_2017_2022['diff'] = diff_mlepen_2017_2022['pvoixMLEPEN'] - diff_mlepen_2017_2022['pvoixMLEPEN_2017']
```

```python
diff_mlepen_2017_2022
```

```python
import folium
import branca.colormap as cm

# Centrer la carte sur la France
latitude_france = 46.603354
longitude_france = 1.888334

carte_france = folium.Map(location=[latitude_france, longitude_france], zoom_start=6)

# Créer une échelle de couleur pour la variable 'A'
val_min = 0.5
val_max = -0.5
colormap = cm.LinearColormap(["blue", "white", "red"]).to_step(10).scale(val_max,val_min )

colormap.caption = "Différence Vote M Le Pen 2017-2022"

# Fonction pour styliser chaque polygone
def style_function(feature):
    valeur = feature['properties']['diff']
    return {
        'fillOpacity': 1,
        'weight': 1,
        'color': 'black',
        'fillColor': colormap(valeur) if valeur is not None else 'grey'
    }

# Ajouter les polygones avec style et tooltip, en utilisant prefer_canvas=True
folium.GeoJson(
    diff_mlepen_2017_2022,
    style_function=style_function,
    tooltip=folium.GeoJsonTooltip(
        fields=["nom", "diff"], 
        aliases=["Commune", "Différence"],
        localize=True
    ),
    prefer_canvas=True  # <-- amélioration des performances pour de nombreux polygones
).add_to(carte_france)

# Ajouter la légende
colormap.add_to(carte_france)

# Sauvegarder la carte
carte_france.save("evolution_mlepen_2017_2022.html")
```
